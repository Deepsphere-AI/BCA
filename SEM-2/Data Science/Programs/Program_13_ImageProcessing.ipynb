# **Lab 13: Install, Import OpenCV and Explore an Simple Image for Image Processing**

Here's a step by step guide to install OpenCV, import it in Python, and explore a simple image for image processing:

1. Install OpenCV: You can install OpenCV by running the following command in your terminal: pip install opencv-python.

2. Import OpenCV in Python: To import OpenCV in Python, use the following code: import cv2.

3. Load an Image: To load an image in OpenCV, you can use the following code:

Import CV2

```python
#Importing cv2
import cv2
#importing cv2_imshow from google.colab.patches
from google.colab.patches import cv2_imshow  
from skimage import color
from skimage import io                         

```

Reading the image file

```python
#Reading the image
#Replace '/content/images.jpg' with your image
vAR_img = cv2.imread('/content/images.jpg', cv2.IMREAD_UNCHANGED)          
```

Displaying the image

```python
#Displaying the image
print("**************************OUTPUT***************************** \n")

cv2_imshow(vAR_img)

print("\n*************************************************************")
```

```python
#converting image from color to gray
vAR_gray_image = cv2.cvtColor(vAR_img, cv2.COLOR_BGR2GRAY)
print("**************************OUTPUT***************************** \n")
cv2_imshow(vAR_gray_image)
print("\n*************************************************************")
```

```python
#Applying the Threshold on image
vAR_ret, vAR_thresh1 = cv2.threshold(vAR_gray_image, 120, 255, cv2.THRESH_BINARY)
print("**************************OUTPUT***************************** \n")
cv2_imshow(vAR_thresh1)
print("\n*************************************************************")
```

Shape of the image

```python
#Printing the shape of the image

print("**************************OUTPUT***************************** \n")

print("Image Shape:", vAR_img.shape)

print("\n*************************************************************")
```

Data type of the image

```python
#Printing the data type of the image

print("**************************OUTPUT***************************** \n")

print("Image Data Type:", vAR_img.dtype)

print("\n*************************************************************")
```
