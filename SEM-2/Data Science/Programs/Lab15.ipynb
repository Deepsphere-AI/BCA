Line Plot: To create a simple line plot in Matplotlib, you can use the following code:

```python
import matplotlib.pyplot as plt

vAR_x = [1, 2, 3, 4, 5]
vAR_y = [2, 4, 6, 8, 10]
plt.plot(vAR_x, vAR_y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

Bar Plot: To create a simple bar plot in Matplotlib, you can use the following code:

```python
import matplotlib.pyplot as plt

vAR_x = [1, 2, 3, 4, 5]
vAR_y = [2, 4, 6, 8, 10]
plt.bar(vAR_x, vAR_y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Plot')
plt.show()
```

Scatter Plot: To create a simple scatter plot in Matplotlib, you can use the following code:

```python
import matplotlib.pyplot as plt

vAR_x = [1, 2, 3, 4, 5]
vAR_y = [2, 4, 6, 8, 10]
plt.scatter(vAR_x, vAR_y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

Histogram: To create a simple histogram in Matplotlib, you can use the following code:

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(100, 10, 1000)
plt.hist(data, bins=30)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram')
plt.show()
```

Pie Chart: To create a simple pie chart in Matplotlib, you can use the following code:

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
values = [20, 30, 40, 10]
plt.pie(values, labels=labels, startangle=90, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Pie Chart')
plt.show()
```

Box Plot:

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(100, 10, 1000)
plt.boxplot(data)
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```

Contour plot:

```python
import matplotlib.pyplot as plt
import numpy as np

def f(vAR_x, vAR_y):
    return np.sin(vAR_x) + np.cos(vAR_y)

vAR_x = np.linspace(-3, 3, 100)
vAR_y = np.linspace(-3, 3, 100)
vAR_X, vAR_Y = np.meshgrid(vAR_x, vAR_y)
vAR_Z = f(vAR_X, vAR_Y)

plt.contour(vAR_X, vAR_Y, vAR_Z)
plt.show()
```
