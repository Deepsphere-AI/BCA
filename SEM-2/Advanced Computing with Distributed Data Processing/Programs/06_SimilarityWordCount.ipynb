```python
# ***********************************************************************************************
  
  # File Name 	:   06_SimilarityWordCount.ipynb
  # Purpose 	  :   Create an application using Apache Spark for Similarity word count 
  # Author	    :   Deepsphere.ai
  # Reviewer 	  :   Jothi Periasamy
  # Date        :   08/02/2023 
  # Version	    :   1.0	
  
# ***********************************************************************************************

## Program Description : The objective of program is counting similarity word in the given text file using pyspark.

## Python Development Environment & Runtime - Python, Anaconda

# ***********************************************************************************************

# installation for pyspark
!pip install pyspark

from pyspark import SparkConf, SparkContext
import urllib.request

# Set up the Spark configuration 

conf = SparkConf().setAppName("SimilarityWordCount")
sc = SparkContext(conf=conf)

# Read in the documents as RDDs
path="https://raw.githubusercontent.com/Deepsphere-AI/BCA/main/SEM-2/Advanced%20Computing%20with%20Distributed%20Data%20Processing/Programs/content/demo.txt"

urllib.request.urlretrieve(path,"demo.txt")

with open("/content/demo.txt","r") as fi:
  content=fi.read()
  punch=[".",","]
  for i in punch:
    for j in content:
      if i == j:
        content=content.replace(i,"")

with open("/content/input.txt","w") as input:
  input.write(content)


doc1 = sc.textFile("/content/input.txt")

# Perform word count on the documents
vAR_wordCount1 = doc1.flatMap(lambda line: line.split(" ")) \
                .map(lambda word: (word, 1)) \
                .reduceByKey(lambda a, b: a + b)

# Print the result
vAR_a=vAR_wordCount1.collect()
vAR_unique_word_count=len(vAR_a)
print("\n******************************** OUTPUT ****************************************\n")

print("Unique word count : ",vAR_unique_word_count ,"\n\n")

print("SimilarityWordCount : \n")
for i in vAR_a:
  print(i[0]+" => ",i[1])

print("************************************************************************")
sc.stop()
```
